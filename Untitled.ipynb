```python
import pandas as pd
```

**Pandas DataFrame** — это двумерная структура данных, в которой данные выровнены в табличной форме с разными строками и столбцами.

Датафрейм можно создавать из объектов Python, например, из словарей. Ключ - название колонки, а значения - список значений для данной колонки.

```python
df2 = pd.DataFrame.from_dict({'a':[1,2],'b':[3,4]})
df2
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/jorisvandenbossche/pandas-tutorial/master/data/titanic.csv')
```

### **info()** - выдает общую информацию об объекте (количество записей, описание всех колонок, описание, сколько колонок не пустые).

```python
df.info()
```
